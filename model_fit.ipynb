```python
import tensorflow as tf
```

```python
pretrained_model=tf.keras.applications.MobileNetV3Large(input_shape=(224,224,3),include_top=False,weights='imagenet',dropout_rate=0.7)
```

```python
pretrained_model.summary()
```

```python
for layer in pretrained_model.layers:
    layer.trainable=True
```

```python
ol=pretrained_model.layers[-1].output
X=pretrained_model.layers[0].input
flatten=tf.keras.layers.Flatten()(ol)
fc_layer=tf.keras.layers.Dense(128)(flatten)
Y=tf.keras.layers.Dense(9,activation='softmax')(fc_layer)
model=tf.keras.Model(X,Y)
```

```python
model.summary()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_gen= ImageDataGenerator()
train_flow_gen=train_gen.flow_from_directory(directory='train_resize',
                                            target_size=(224,224),
                                            class_mode='categorical',
                                            batch_size=64,
                                            shuffle=True)
```

```python
# 30분 넘게 걸림
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(train_flow_gen,epochs=1)
```
